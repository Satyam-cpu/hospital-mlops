```python
df.columns = df.columns.str.strip()
df.columns.tolist()
```

```python
df.isna().sum()
```

```python
df["Average Total Payments"] = (
    df["Average Total Payments"]
    .str.replace("$","",regex=False)
    .str.replace(",","",regex=False)
    .astype(float)
)
```

```python
money_cols = [
    "Average Covered Charges",
    "Average Medicare Payments"
]

for col in money_cols:
    df[col] = (
        df[col]
        .str.replace("$","",regex=False)
        .str.replace(",","",regex=False)
        .astype(float)
    )
```

```python
features = [
    "Provider State",
    "DRG Definition",
    "Total Discharges",
    "Average Covered Charges"
]

X = df[features]
y = df["Average Total Payments"]
```
